
###  **Theory Part**

1. **What is hypothesis testing in statistics?**
-  Hypothesis testing is a method used in statistics to make decisions based on data. It helps determine whether a certain assumption (hypothesis) about a population parameter is true or not using sample data.

2. **What is the null hypothesis, and how does it differ from the alternative hypothesis?**
- The null hypothesis (H₀) assumes there is no effect or difference, while the alternative hypothesis (H₁) suggests that there is an effect or difference. We test the null and try to find evidence to reject it.

3. **What is the significance level in hypothesis testing, and why is it important?**
- The significance level (α) is the probability of rejecting a true null hypothesis. Common values are 0.05 or 5%. It helps decide the threshold for whether the results are statistically significant.

4. **What does a P-value represent in hypothesis testing?**
- The P-value shows the probability of obtaining the observed data (or something more extreme) if the null hypothesis is true.

5. **How do you interpret the P-value in hypothesis testing?**
- If the P-value is less than the significance level (e.g., 0.05), we reject the null hypothesis. If it's greater, we fail to reject it.

6. **What are Type 1 and Type 2 errors in hypothesis testing?**

   * Type 1 Error: Rejecting the null when it is actually true.
   * Type 2 Error: Not rejecting the null when it is actually false.

7. **What is the difference between a one-tailed and a two-tailed test?**
- A one-tailed test checks for a difference in one direction (greater or less), while a two-tailed test checks for any difference (both directions).

8. **What is the Z-test, and when is it used?**
- Z-test is used to test if there's a significant difference between sample and population means when the population standard deviation is known.

9. **How do you calculate the Z-score, and what does it represent?**
- Z = (X - μ) / σ
   It shows how many standard deviations a value is from the mean.

10. **What is the T-distribution, and when should it be used instead of the normal distribution?**
- T-distribution is used when the sample size is small and the population standard deviation is unknown.

11. **What is the difference between a Z-test and a T-test?**
- Z-test requires known population variance, while T-test is used when it's unknown and sample size is small.

12. **What is the T-test, and how is it used?**
- T-test compares sample means to see if they come from populations with the same mean.

13. **What is the relationship between Z-test and T-test?**
- Both test means, but Z-test is for large samples or known σ, while T-test is for small samples or unknown σ.

14. **What is a confidence interval, and how is it used?**
- It is a range of values that is likely to contain the population parameter. It helps understand the precision of our estimate.

15. **What is the margin of error, and how does it affect the confidence interval?**
- Margin of error is the range added and subtracted from the sample estimate to create a confidence interval. Higher margin means wider interval.

16. **How is Bayes' Theorem used in statistics, and what is its significance?**
- Bayes' Theorem helps update probability estimates based on new evidence. It’s key in decision-making and inference.

17. **What is the Chi-square distribution, and when is it used?**
- It’s used for categorical data analysis, especially in tests like goodness-of-fit and independence.

18. **What is the Chi-square goodness of fit test?**
- It checks whether a sample matches an expected distribution.

19. **What is the F-distribution, and when is it used?**
- It is used in ANOVA to compare variances between groups.

20. **What is an ANOVA test, and what are its assumptions?**
- ANOVA (Analysis of Variance) compares means across 3 or more groups assuming independence, normality, and equal variances.

21. **What are the different types of ANOVA tests?**

    * One-way ANOVA
    * Two-way ANOVA
    * Repeated measures ANOVA

22. **What is the F-test, and how does it relate to hypothesis testing?**
- The F-test compares variances and is the basis for ANOVA. It tests if group variances are significantly different.

```python
#4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.random.normal(51, 4, 40)
pop_mean = 50
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

z = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

critical_value = norm.ppf(0.975)

plt.plot(x, y, label='Standard Normal Curve')
plt.fill_between(x, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.3, label='Rejection Region')
plt.axvline(z, color='blue', linestyle='--', label=f'Z = {z:.2f}')
plt.title("Two-tailed Z-Test")
plt.legend()
plt.grid(True)
plt.show()

print("Z-Statistic:", z)
print("P-value:", p_value)
```

```python
#5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=50, mu1=52, sigma=5, n=30, alpha=0.05):
    se = sigma / np.sqrt(n)
    z_critical = norm.ppf(1 - alpha / 2)
    lower_bound = mu0 - z_critical * se
    upper_bound = mu0 + z_critical * se

    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    plt.plot(x, y0, label='H0 Distribution', color='blue')
    plt.plot(x, y1, label='H1 Distribution', color='green')

    plt.fill_between(x, y0, where=(x <= lower_bound) | (x >= upper_bound), color='red', alpha=0.3, label='Type I Error (α)')
    plt.fill_between(x, y1, where=(x > lower_bound) & (x < upper_bound), color='orange', alpha=0.3, label='Type II Error (β)')

    plt.axvline(lower_bound, color='black', linestyle='--')
    plt.axvline(upper_bound, color='black', linestyle='--')
    plt.title("Type I and Type II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

visualize_type1_type2_errors()
```

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)
    interval = stats.t.interval(confidence, len(data)-1, loc=mean, scale=sem)
    return interval

data = np.random.normal(100, 10, 30)
ci = confidence_interval(data)
print("Confidence Interval (95%):", ci)
```

```python
#11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayes_theorem(prior_A, prob_B_given_A, prob_B):
    return (prob_B_given_A * prior_A) / prob_B

# Example values
prior_A = 0.01            # P(Disease)
prob_B_given_A = 0.99     # P(Positive | Disease)
prob_B_given_not_A = 0.05 # P(Positive | No Disease)
prior_not_A = 1 - prior_A # P(No Disease)

# Total probability of B (positive test)
prob_B = (prob_B_given_A * prior_A) + (prob_B_given_not_A * prior_not_A)

# Apply Bayes' Theorem
posterior_A_given_B = bayes_theorem(prior_A, prob_B_given_A, prob_B)

print("P(Disease | Positive Test):", posterior_A_given_B)
```
